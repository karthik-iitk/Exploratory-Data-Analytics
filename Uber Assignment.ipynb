```python
#Importing all the required packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
uber = pd.read_csv('Uber Request Data.csv')
```

```python
#here i have dropped the duplicates if any

uber=pd.DataFrame(uber)
uber=uber.drop_duplicates()
uber
```

```python
#here i have checked the number of NA values in each of the attributes

uber.isna().sum()
```

```python
#Data cleaning 

uber['Drop timestamp']
```

```python
#Data Cleaning

#here i have dropped the Request id and Driver id as they are also of no use for our analysis

uber.drop('Request id', inplace=True, axis=1)
uber.drop('Driver id', inplace=True, axis=1)
uber
```

```python
#Data Cleaning

#hereby i have converted both the Request timestamp and Drop timestamp into Datetime format for uniformity of data

uber['Request timestamp'] = uber['Request timestamp'].apply(lambda x : pd.to_datetime(str(x)))
uber['Drop timestamp'] = uber['Drop timestamp'].apply(lambda x : pd.to_datetime(str(x)))
uber['Request timestamp']=pd.to_datetime(uber['Request timestamp'],dayfirst=True)
uber['Drop timestamp']=pd.to_datetime(uber['Drop timestamp'],dayfirst=True)
uber

```

```python
#checking the datatypes

uber.dtypes

```

```python
#Data Pre-Processing

#creating a new coloumn for better understanding

uber['Request hour']=uber['Request timestamp'].dt.hour
uber
```

```python
#checking the datatypes now and then :P

uber.dtypes
```

```python
#Data Pre-Processing

#creating new coloumns Completed,Cancelled,NoCars for good bivariate analysis.

uber['Completed']=uber['Status']=="Trip Completed"
uber['Cancelled']=uber['Status']=="Cancelled"
uber['NoCars']=uber['Status']=="No Cars Available"

uber
```

```python
#Data Pre-Processing

#here, i have binned the Request hour coloumn in 6 bins so as to ease the analysis in future

uber['Category']=pd.cut(uber['Request hour'], 6,  labels=["LateNight","EarlyMorning","Morning","Afternoon","Evening","Night"], include_lowest=True)
uber
```

```python
#univariate analysis

#to understand the busiest hours of bookings

sns.distplot(uber['Request hour'],rug=True)
plot.show()

# it is pretty evident that the peak hours are at 6AM in the morning and at 6PM in the evening. 

# Maybe its because of the fact that most of the flights land or takeoff from the airport in the analysis at these times.
```

```python
#here i have bin plotted the same thing

#What we infer from here is that the bookings are maximum at Early morning and Evening slots as expected

sns.distplot(uber['Request hour'], bins = 6)
plot.show()
```

```python
sns.boxplot(y=uber['Request hour'])
plot.show()

# a boxplot which clearly shows the 25th percentile,mean and the 75th percentile. It also shows that there are no outliers.
```

```python
#Supply Demand Gaps

#here i have created a new coloumn demand which includes all the trips irrespective of the status.

uber['Demand']=1
uber
```

```python
#Supply Demand Gaps

#here i have created a separate attribute Supply which takes the value 1 if the Demand is met or else 0

uber['Supply'] = uber['Status'].apply(lambda x: 1 if x=='Trip Completed' else 0)
uber
```

```python
#Supply Demand Gaps

#here i am creating a separate coloumn SupplyDemand to find the Supply Demand gap easily

uber['SupplyDemand']=uber['Demand']-uber['Supply']
uber
```

```python
#Supply Demand Gaps

#the below count plot shows the number of demands met and the number of demands unmet.

#they are 2800 and 3900 respectively

sns.countplot(x="SupplyDemand",data=uber)
```

```python
#Supply Demand Gaps

#the below barplot clearly shows the Supply Demand gap is high at LateNight and Evening. And its also varying with categorys

sns.barplot(x="Category", y="SupplyDemand",data=uber)
```
